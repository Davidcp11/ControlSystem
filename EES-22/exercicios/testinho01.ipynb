```python
import numpy as np
import control as ct
import matplotlib.pyplot as plt
import sympy as sp
from sympy import *
```

```python
G1 = ct.tf(np.polymul([1, 1], [6+2*np.sqrt(5)]), [1, 2+2*np.sqrt(5), 6+2*np.sqrt(5)])
t = np.linspace(0,3, 100)
t1, y1 = ct.step_response(G1, t)
plt.plot(t1, y1, 'b-', label='G1')
plt.axhline(1, color='k', linestyle='--')
plt.legend()
```

```python
S = ct.step_info(G1)
for k in S:
    print(f"{k}: {S[k]:.4f}")
```
