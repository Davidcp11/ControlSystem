```python
import numpy as np
import matplotlib.pyplot as plt
import neurokit2 as nk # biblioteca para simular dados de eletrocardiogramas
```

```python
# Gerar um sinal de ECG
signal = nk.ecg_simulate(duration=10, sampling_rate=500)

plt.plot(signal)
plt.title("Sinal Simulado de ECG")
plt.xlabel("Amostras")
plt.ylabel("Amplitude")
plt.show()
```

```python
dados = signal
N = 5000
fs = 500
T = 1/fs
t = np.arange(0, N/fs, T)
```

```python
plt.figure()
plt.plot(t, dados)
```

```python
f = np.fft.fftfreq(N, d=T)
transf = np.fft.fft(dados)
transf = np.abs(transf)
```

```python
plt.plot(f[f>0], transf[f>0])
```
