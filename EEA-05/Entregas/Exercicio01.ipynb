# Constantes
$\omega_p =1$\
$\omega_z =1$\
$Q_p=1$\
$K=1$

# 1) Filtro Passa-baixas 
$H_{LP}(s) = \frac{K \omega_{p}^2}{s^2+(\omega _p/Q_p)s+\omega^2}$\
\
Substituindo os valores\
$H_{LP}(s) = \frac{1}{s^2+s+1}$


```python
Hlp_baixas = ct.tf([1], [1, 1, 1])
print(f"Polos:{ct.poles(Hlp_baixas)}\nZeros:{ct.zeros(Hlp_baixas)}")
responseLP_baixas = ct.pole_zero_map(Hlp_baixas)
ct.pole_zero_plot(responseLP_baixas)
```

```python
ct.bode_plot(Hlp_baixas)
```

# 2) Filtro Passa-altas
$H_{LP}(s) = \frac{K s^2}{s^2+(\omega _p/Q_p)s+\omega^2}$\
\
Substituindo os valores\
$H_{LP}(s) = \frac{s^2}{s^2+s+1}$

```python
Hlp_altas = ct.tf([1, 0, 0], [1, 1, 1])
print(f"Polos:{ct.poles(Hlp_altas)}\nZeros:{ct.zeros(Hlp_altas)}")
responseHlp_altas = ct.pole_zero_map(Hlp_altas)
ct.pole_zero_plot(responseHlp_altas)
```

```python
ct.bode_plot(Hlp_altas)
```

# 3) Filtro Passa-faixas
$H_{LP}(s) = \frac{K(\omega_p/Q_p) s}{s^2+(\omega _p/Q_p)s+\omega^2}$\
\
Substituindo os valores\
$H_{LP}(s) = \frac{s}{s^2+s+1}$

```python
Hlp_faixas = ct.tf([1, 0], [1, 1, 1])
print(f"Polos:{ct.poles(Hlp_faixas)}\nZeros:{ct.zeros(Hlp_faixas)}")
responseLP_faixas = ct.pole_zero_map(Hlp_faixas)
ct.pole_zero_plot(responseLP_faixas)
```

```python
ct.bode_plot(Hlp_faixas)
```

# 4) Filtro Nortch
$H_{LP}(s) = \frac{K(s^2+\omega_z^2)}{s^2+(\omega_p/Q_p)s+\omega^2}$\
\
Substituindo os valores\
$H_{LP}(s) = \frac{s^2+1}{s^2+s+1}$

```python
Hlp_nortch = ct.tf([1, 0, 1], [1, 1, 1])
print(f"Polos: {ct.poles(Hlp_nortch)}\nZeros: {ct.zeros(Hlp_nortch)}")
responseLP_nortch = ct.pole_zero_map(Hlp_nortch)
ct.pole_zero_plot(responseLP_nortch)
```

```python
ct.bode_plot(Hlp_nortch)
```

# 5) Filtro Passa-tudo
$H_{BP}(s) = K\frac{s^2-(\omega_p/Q_p)s+\omega_z ^2}{s^2+(\omega _p/Q_p)s+\omega^2}$\
\
Substituindo os valores\
$H_{BP}(s) = \frac{s^2-s+1}{s^2+s+1}$

```python
Hbp_tudo = ct.tf([1, -1, 1], [1, 1, 1])
print(f"Polos:{ct.poles(Hbp_tudo)}\nZeros: {ct.zeros(Hbp_tudo)}")
responseHbp_tudo = ct.pole_zero_map(Hbp_tudo)
ct.pole_zero_plot(responseHbp_tudo)
```

```python
ct.bode_plot(Hbp_tudo)
```
